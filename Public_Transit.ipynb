```python
import pandas as pd
import numpy as np
import googlemaps
from datetime import datetime
gmaps = googlemaps.Client(key='AIzaSyDA1tcl8xYJGOwam-Ut5VPEn5T5lRRqAac')
df_main = pd.read_csv('NYC_Transit_Subway_Entrance_And_Exit_Data.csv')
'''此py将站台线路信息表转化为neighbr线路数量表'''
```

```python
def bool_route(x): #用来将原表格的route信息转换为每个站台有几条线
    t = type(x)
    if t == str:
        return 1
    return 0
df_main['Total_Number'] = df_main['Route1'].map(bool_route)+df_main['Route2'].map(bool_route)+\
                        df_main['Route3'].map(bool_route)+df_main['Route4'].map(bool_route)+\
                        df_main['Route5'].map(bool_route)+df_main['Route6'].map(bool_route)+\
                        df_main['Route7'].map(bool_route)+df_main['Route8'].map(bool_route)+\
                        df_main['Route9'].map(bool_route)+df_main['Route10'].map(bool_route)+\
                        df_main['Route11'].map(bool_route)
```

```python
def get_nghbr(loc):#用googlemap api 获得站台所在neighbor
    loc = eval(loc)
    try:
        reverse_geocode_result = gmaps.reverse_geocode(loc,result_type='neighborhood')
        neghbr = reverse_geocode_result[0]['address_components'][0]['long_name'].upper()
        return neghbr
    except:
        return None
```

```python
df_main['neighbhood'] = df_main['Station Location'].apply(get_nghbr)
```

```python
df_re = df_main[['Total_Number','neighbhood']].copy()
```

```python
df_final = df_re.groupby(df_re['neighbhood']).max() #用neighbr给站台分组，并取出每组线路的最大值作为该neighbr的线路数
```
